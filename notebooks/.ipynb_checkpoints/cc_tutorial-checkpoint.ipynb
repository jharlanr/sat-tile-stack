```python
## ======== ##
## PLOTTING ##
## ======== ##

def linear_stretch_rgb(rgb, pmin=2, pmax=98):
    out = np.empty_like(rgb)
    for i in range(3):
        band = rgb[..., i]
        lo, hi = np.nanpercentile(band, (pmin, pmax))
        out[..., i] = np.clip((band - lo) / (hi - lo), 0, 1)
    return out

def plot_n_days(timestack, start_date, ndays=10, cols=4):
    """
    Plot 10 consecutive daily RGB composites in subplots.

    Parameters
    ----------
    timestack : xarray.DataArray
        DataArray of shape (time, band, y, x).
    start_date : str or numpy.datetime64
        The first date to plot, e.g. "2019-05-20".
    cols : int
        Number of columns in the grid (default 4).
    """
    # Ensure start_date is numpy.datetime64
    start = np.datetime64(start_date)

    n_days = ndays
    rows = int(np.ceil(n_days / cols))

    fig, axes = plt.subplots(
        rows, cols,
        figsize=(cols * 3, rows * 3),
        constrained_layout=True
    )

    axes_flat = axes.flatten()

    for i in range(n_days):
        date = start + np.timedelta64(i, 'D')
        tslice = timestack.sel(time=date)
        # build and stretch RGB
        r = tslice.sel(band="B04").values
        g = tslice.sel(band="B03").values
        b = tslice.sel(band="B02").values
        rgb = np.stack([r, g, b], axis=-1)
        rgb_st = linear_stretch_rgb(rgb, pmin=0, pmax=95)

        ax = axes_flat[i]
        ax.imshow(rgb_st)
        ax.set_title(str(date)[:10])
        ax.axis("off")

    # Remove any unused axes
    for j in range(n_days, rows * cols):
        fig.delaxes(axes_flat[j])

    plt.show()
    
plot_n_days(timestack, "2019-08-03", ndays=20, cols=5)

```
